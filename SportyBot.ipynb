```python
from aiogram import Bot, Dispatcher, types, F
import asyncio
from aiogram.filters import CommandStart, Command, or_f
from aiogram.types import ReplyKeyboardMarkup, KeyboardButton, ReplyKeyboardRemove, BotCommand
import random


# Мужчины: видео для упражнений для новичков
male_arm_workout_video_beginner = 'https://www.youtube.com/watch?v=J1-SDv_kwiY'  # тренировка на руки
male_leg_workout_video_beginner = 'https://youtu.be/xzIbXYX3Fes?si=YB-KdgZziLI7kHrH'  # тренировка на ноги
male_abs_workout_video_beginner = 'https://youtu.be/3p8EBPVZ2Iw?si=m6MYiN0DVE9QLyk7'  # тренировка на пресс
male_whole_body_workout_video_beginner = 'https://youtu.be/iCQ2gC4DqJw?si=xTkJsAtzi7F1J5Et'  # тренировка на все тело

# Мужчины: видео для упражнений для опытных
male_arm_workout_video_advanced = 'https://www.youtube.com/watch?v=wwKb-wZCEjs'  # тренировка на руки
male_leg_workout_video_advanced = 'https://www.youtube.com/watch?v=841gJUczmzg'  # тренировка на ноги
male_abs_workout_video_advanced = 'https://www.youtube.com/watch?v=-hqYWRib5y8'  # тренировка на пресс
male_whole_body_workout_video_advanced = 'https://www.youtube.com/watch?v=wk4-TDsTF_8'  # тренировка на все тело


# Женщины: видео для упражнений для новичков
female_arm_workout_video_beginner = 'https://youtu.be/8qDDtm6BOfw?si=ECpc86d14Envogne'  # тренировка на руки
female_leg_workout_video_beginner = 'https://youtu.be/nG69wuXHwxg?si=YXr18Y_HpGRGaFE5'  # тренировка на ноги
female_abs_workout_video_beginner = 'https://youtu.be/1f8yoFFdkcY?si=OLO7u33HTaDmz0VV'  # тренировка на пресс
female_whole_body_workout_video_beginner = 'https://youtu.be/nT_pwsGNn44?si=lhAh3Ah1ym9CVZ4n'  # тренировка на все тело

# Женщины: видео для упражнений для опытных
female_arm_workout_video_advanced = 'https://youtu.be/Y346900i9qE?si=LOE7QHN4_hGSfb3X'  # тренировка на руки
female_leg_workout_video_advanced = 'https://youtu.be/uVwNVEQS_uo?si=AC_c5l7JTlxG4sF0'  # тренировка на ноги
female_abs_workout_video_advanced = 'https://youtu.be/hbjq9_NER_k?si=lpqcW1Sla45FN-Mh'  # тренировка на пресс
female_whole_body_workout_video_advanced = 'https://youtu.be/l9_SoClAO5g?si=QIVWIGSIA9TeGUjL'  # тренировка на все тело

# Словарь - мужчины: упражнения для новичков
male_light_exercises = {
    "arms": {  # Упражнения на руки
        "PIKES": "pending",
        "PIKE PUSH-UPS": "pending",
        "DIAMOND PUSH-UPS": "pending",
        "BENCH DIPS": "pending",
        "PLANK-UPS": "pending",
        "SKULL CRUSHERS": "pending",
        "BODY WEIGHT ROW": "pending",
        "BICEP CURLS": "pending",
        "FINGER PRESS": "pending"
    },
    "legs": {  # Упражнения на ноги
        "SQUATS": "pending",
        "BULGARIAN SPLIT SQUATS": "pending",
        "CROSSACK SQUAT": "pending",
        "JUMPING LUNGES": "pending",
        "SIDE LEG ABDUCTION": "pending",
        "EXPLOSIVE SQUATS": "pending",
        "SINGLE LEG DEADLIFT": "pending",
        "SINGLE LEG GLUTE BRIDGE": "pending"
    },
    "abs": {  # Упражнения на пресс
        "HIGH KNEE TAPS": "pending",
        "RUSSIAN TWISTS": "pending",
        "LEG RAISES": "pending",
        "HIP RAISES": "pending",
        "FLUTTER KICKS": "pending",
        "PLANK KNEES TO ELBOW": "pending",
        "CHAIR SIT UPS": "pending",
        "SEATED IN AND OUTS": "pending",
        "JUMPING JACKS": "pending"
    },
    "whole_body": {  # Упражнения на все тело
        "SQUATS": "pending",
        "SINGLE LEG RAISE": "pending",
        "IN AND OUT SQUATS": "pending",
        "CRUNCHES WITH LEGS DOWN": "pending",
        "PUSH UPS": "pending",
        "SIDE SQUATS": "pending",
        "CROSS CRUNCHES": "pending",
        "MOUNTAIN CLIMBER KNEE TO ELBOW": "pending"
    }
}

# Словарь - мужчины: упражнения для опытных
male_advanced_exercises = {
    "arms": {  # Упражнения на руки
        "DIAMOND PUSH-UPS": "pending",
        "TOWEL ONE LEG BICEP CURLS": "pending",
        "IN AND OUT PUSH-UPS": "pending",
        "TWO LEG BICEPS CURLS 1": "pending",
        "DRAGON": "pending",
        "TWO LEG BICEPS CURLS 2": "pending",
        "SLOW PUSH-UPS": "pending",
        "TWO LEG BICEPS CURLS 3": "pending"
    },
    "legs": {  # Упражнения на ноги
        "DOUBLE SQUAT": "pending",
        "SQUATS IN ONE LEG": "pending",
        "SQUATS IN ANOTHER LEG": "pending",
        "BULGARIAN SQUATS NON STOP": "pending",
        "BULGARIAN SQUATS NON STOP ANOTHER LEG": "pending",
        "3 WAY SQUAT": "pending",
        "3 WAY DOWN SQUAT LOW": "pending",
        "JUMP SQUAT": "pending",
        "JUMP SQUAT AND PAUSE": "pending"
    },
    "abs": {  # Упражнения на пресс
        "STARFISH CRUNCH": "pending",
        "SIDE KNEE TO ELBOW": "pending",
        "CRAB LEG RAISE": "pending",
        "LEG SCISSOR": "pending",
        "DYNAMIC PLANK": "pending",
        "OBLIQUE HEEL TAPS": "pending",
        "FLYING KNEE RAISE": "pending",
        "SEATED ABS CIRCLE": "pending",
        "BICYCLE KICKS": "pending"
    },
    "whole_body": {  # Упражнения на все тело
        "HIGH KNEES": "pending",
        "JUMP SQUAT": "pending",
        "BURPEE PUSH UP": "pending",
        "EXTENDED CRUNCH": "pending",
        "JUMPING JACKS": "pending",
        "SUPERMAN PULL UP": "pending",
        "MOUNTAIN CLUMBERS": "pending",
        "BUT KICKS": "pending",
        "PLANK REACHES": "pending"
    }
}

# Словарь - женщины: упражнения для новичков
female_light_exercises = {
    "arms": {  # Упражнения на руки
        "ARMS CROSS OVER": "pending",
        "ARMS UP AND DOWN": "pending",
        "TRICEP PRESS": "pending",
        "PALM DOWN": "pending",
        "ARM WING": "pending"
    },
    "legs": {  # Упражнения на ноги
        "SLOW SQUAT": "pending",
        "SIDE STEP SQUAT": "pending",
        "SQUAT + SIDE LEG LIFT": "pending",
        "KNEELING SQUAT": "pending",
        "KNEELING SQUAT + 3 PULSES": "pending",
        "KNEELING LEAN BACK": "pending",
        "INNER THIGH LIFT LEFT": "pending",
        "INNER THIGH LIFT HOLD 1": "pending",
        "INNER THIGH LIFT RIGHT": "pending",
        "INNER THIGH LIFT HOLD 2": "pending",
        "GLUTE BRIDGE": "pending",
        "WIDE GLUTE BRIDGE PULSES": "pending",
        "GLUTE BRIDGE HOLD": "pending",
        "1-LEG FROG KICKS": "pending",
        "FROG HOLD": "pending",
    },
    "abs": {  # Упражнения на пресс
        "CRUNCH MIDDLE": "pending",
        "CRUNCH LEFT": "pending",
        "CRUNCH RIGHT": "pending",
        "HEEL TOUCHES": "pending",
        "AB HOLD 1": "pending",
        "RUSSIAN TWIS": "pending",
        "LEAN BACK HOLD": "pending",
        "SLOW MOUNTAIN CLIMBER": "pending",
        "PLANK 1": "pending",
        "TOE TOUCHES": "pending",
        "BICYCLE CRUNCH": "pending",
        "HIP LIFT": "pending",
        "AB HOLD  2": "pending",
        "SPIDER PLANK": "pending",
        "PLANK 2": "pending",
    },
    "whole_body": {  # Упражнения на все тело
        "QUICK FEET": "pending",
        "SKIPPING": "pending",
        "BUTT KICKS": "pending",
        "SQUATS": "pending",
        "LUNGES": "pending",
        "PUSH-UPS": "pending",
        "PLANKS": "pending",
        "SIT-UPS": "pending"
    }
}

# Словарь - женщины: упражнения для опытных
female_advanced_exercises = {
    "arms": {  # Упражнения на руки
        "GOLF BALLS": "pending",
        "WALNUT CRUSHERS": "pending",
        "PRAYER PULSES": "pending",
        "BACK BEHINDS": "pending",
        "ANGEL ARMS": "pending",
        "PALMS BACK PULSES": "pending",
        "TRIANGLE PUSHES": "pending",
        "ROBOT ARMS": "pending",
        "GOAL POST PUNCHERS": "pending",
        "ROOF RAISERS": "pending"
    },
    "legs": {  # Упражнения на ноги
        "DIAGONAL FORWARD ALTERNATING LUNGE": "pending",
        "SIDE TO SIDE": "pending",
        "REVERSE ALTERNATING": "pending",
        "LUNGE TO ROTATIONAL KNEE": "pending",
        "FORWARD ALTERNATING LUNGE": "pending",
        "FORWARD ALTERNATING LUNGE W/ 2 PULSES ": "pending",
        "SIDE LUNGE TO CURTSEY": "pending",
        "PULSE LUNGES": "pending"

    },
    "abs": {  # Упражнения на пресс
        "CRUNCH": "pending",
        "CRUNCH WITH KNEE LIFT": "pending",
        "BICYCLE CRUNCH 1": "pending",
        "TOE REACH CRUNCH": "pending",
        "CRUNCH WITH LEG LIFT": "pending",
        "REACH OUT - TUCK IN 1": "pending",
        "ROLL IN 1": "pending",
        "BICYCLE CRUNCH 2": "pending",
        "OBLIQUE CRUNCH (RIGHT) 1": "pending",
        "OBLIQUE CRUNCH (LEFT) 1": "pending",
        "SITTING CRUNCH": "pending",
        "OBLIQUE CRUNCH (RIGHT) 2": "pending",
        "OBLIQUE CRUNCH (LEFT) 2": "pending",
        "JACK KNIFE": "pending",
        "STARFISH CRUNCH": "pending",
        "REACH OUT - TUCK IN 2": "pending",
        "STARFISH TO JACK KNIFE": "pending",
        "ROLL IN 2": "pending",
        "HOLD": "pending",
        "PLANK": "pending"

    },
    "whole_body": {  # Упражнения на все тело
        "CHEST PRESS": "pending",
        "ALTERNATING CHEST PRESS": "pending",
        "DEADSTOP CHEST PRESS": "pending",
        "SUITCASE SQUAT": "pending",
        "SLOW LOWER GOBLET SQUAT": "pending",
        "1/2 REP SQUATS": "pending",
        "RENEGADE ROW": "pending",
        "ALTERNATING RENEGADE ROW": "pending",
        "STATIC LUNGE": "pending",
        "FORWARD LEAN LUNGE": "pending",
        "SHOULDER PRESS": "pending",
        "PUSH PRESS": "pending",
        "ARNOLD PRESS": "pending",
        "RDL SLOW": "pending",
        "RDL PAUSE AT BOTTOM": "pending",
        "1/2 REP RDL": "pending"
    }
}

# Множество мотивационных цитат
with_skip = {'"Даже самый длинный путь начинается с первого шага." - Лао-цзы',
               'Победа не всегда будет приходить легко. Но с каждым упорным усилием вы приближаетесь к своей цели!',
               '"Начинайте там, где вы находитесь. Используйте то, что у вас есть. Делайте, что вы можете." '
               '- Артур Эш',
               'Секрет успеха - не бояться провалов на пути к своим целям',
               'Не сравнивайте свой прогресс с прогрессом других. Сравнивайте с тем, какой были вы вчера',
               'То, что сегодня кажется невозможным, завтра станет вашим достижением',
               'Верьте в себя. Вы способны на большее, чем вы думаете',
               'Каждый день - новая возможность стать лучше, чем вы были вчера',
               'Путь к успеху - это последовательность маленьких шагов, делаемых ежедневно',
               'Помните, что каждый профессионал был когда-то новичком. Не стесняйтесь своего начального уровня',
               'Победа - это не финальная точка, это часть пути. Наслаждайтесь каждым шагом',
               'Самое сложное - это начать. Но только начав, вы уже на полпути к своей цели',
               'Помните, что ваша цель стоит всех усилий и труда, который вы вкладываете в нее',
               'Самый важный шаг - это тот, который вы делаете сейчас',
               'Каждый раз, когда вы заканчиваете тренировку, вы становитесь сильнее, чем были раньше',
               'Ваша постоянная преданность тренировкам приведет вас к результатам, которых вы желаете',
               'Сегодняшняя тренировка - это ваш следующий шаг к лучшей версии себя',
               'Каждая завершенная тренировка приближает вас к вашим целям. Вы на правильном пути!',
               'Сегодняшний успех - это результат ваших вчерашних усилий. Продолжайте в том же духе!',
               'Вы доказали себе, что сможете преодолеть любые препятствия на своем пути к успеху',
               'Ваша решимость и настойчивость важнее, чем любые ограничения, которые могут вас задержать',
               'Завершив тренировку, вы дарите себе не только физическое укрепление, но и душевное удовлетворение',
               'С каждой законченной тренировкой вы становитесь все более решительным и уверенным в своих силах',
               'Ваш труд и самоотверженность превращают ваши мечты в реальность. Продолжайте двигаться вперед!',
               'Не существует ничего более вдохновляющего, чем видеть, как вы достигаете своих целей через свои '
               'усилия',
               'Сегодняшняя тренировка - это еще один кирпичик в построении вашего здорового и сильного тела',
               'Позвольте себе насладиться своими достижениями. Вы сделали отличную работу, и ваш труд не прошел '
               'даром!'}


TOKEN = '7051179290:AAEyKh2aL8ZKsyIkyeP2-xuIXVAE4cHWdcE'  # Токен телеграмм-бота

bot = Bot(TOKEN)  # Создание экземпляра бота с использованием предоставленного токена
dp = Dispatcher()  # Создание диспетчера для обработки входящих сообщений и команд


del_kbd = ReplyKeyboardRemove()  # Удаление кнопок после нажатия

picked_workout = {}  # Словарь для определения тренировки

completed_all_exercises = False  # Инициализация переменной, указывающей на то, что все упражнения еще не завершены


async def set_commands(bot: Bot):  # Список команд через /
    commands = [BotCommand(command='start', description='Начать'),
                BotCommand(command='workouts', description='Тренировки'),
                BotCommand(command='motivation', description='Заряд мотивации')]
    await bot.set_my_commands(commands)


async def main():
    await set_commands(bot)  # Активация команд
    await dp.start_polling(bot)  # Запуск бота


kb_experience = ReplyKeyboardMarkup(  # Кнопки для выбора наличия/отсутствия опыта пользователя
    keyboard=[
        [
            KeyboardButton(text='Без опыта'),
            KeyboardButton(text='С опытом')
        ]
    ],
    resize_keyboard=True
)

kb_type = ReplyKeyboardMarkup(  # Кнопки для выбора типа тренировок
    keyboard=[
        [
            KeyboardButton(text='На руки'),
            KeyboardButton(text='На ноги')
        ],
        [
            KeyboardButton(text='На пресс'),
            KeyboardButton(text='На все тело'),
        ],
    ],
    resize_keyboard=True
)

kb_complete_skip = ReplyKeyboardMarkup(  # Кнопки для выполнения/пропуска упражнения
    keyboard=[
        [
            KeyboardButton(text='Выполнено'),
            KeyboardButton(text='Пропустить')
        ]
    ],
    resize_keyboard=True
)

kb_gender = ReplyKeyboardMarkup(  # Кнопки для выбора пола пользователя
    keyboard=[
        [
            KeyboardButton(text='Мужчина'),
            KeyboardButton(text='Женщина')
        ]
    ],
    resize_keyboard=True
)

kb_motivation = ReplyKeyboardMarkup(  # Кнопки для выбора действия с мотивационной цитатой
    keyboard=[
        [
            KeyboardButton(text='Больше мотивации!'),
            KeyboardButton(text='Приступим к тренировке!')
        ],
    ],
    resize_keyboard=True
)


@dp.message(CommandStart())  # Обработка команды /start
async def start_cmd(message: types.Message):
    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    # Приветствие и исчезание кнопок
    await message.answer('Вас приветствует бот, который поможет вам со спортивными тренировками!\n'
                         'Для выбора тренировки воспользуйтесь командой /workouts', reply_markup=del_kbd)


@dp.message(Command("workouts"))  # Обработка команды /workouts
async def handle_workouts_command(message: types.Message):
    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    # Отправление сообщения с выбором наличия/отсутствия опыта
    await message.answer('Выберите ваш опыт тренировок:', reply_markup=kb_experience)


# Обработка нажатия на кнопки с выбором опыта тренировок
@dp.message(lambda message: message.text.lower() in ["без опыта", "с опытом"])
async def handle_experience_choice(message: types.Message):
    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    picked_workout['experience'] = message.text.lower()  # Выбор наличия или отсутствия опыта в тренировках

    # Отправка сообщения с выбором пола пользователя
    await message.answer('Кто вы?', reply_markup=kb_gender)


# Обработка нажатия на кнопки с выбором пола пользователя
@dp.message(lambda message: message.text.lower() in ["мужчина", "женщина"])
async def handle_experience_choice(message: types.Message):
    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    picked_workout['gender'] = message.text.lower()  # Внесение данных в ключ "гендер" пользователя

    # Отправка сообщения с выбором типа тренировки
    await message.answer('Теперь выберите тип тренировки:', reply_markup=kb_type)


# Обработка нажатия на кнопки с выбором типа упражнений
@dp.message(or_f(F.text.lower() == "на руки", F.text.lower() == "на ноги",
                 F.text.lower() == "на пресс", F.text.lower() == "на все тело"))
async def handle_exercise_type(message: types.Message):
    # Глобальные переменные:
    # 1. список упражнений, предназначенных в текущей тренировке;
    # 2. кол-во выполненных упражнений в текущей тренировке,
    # 3. переменная, отслеживающая были ли выполнены все упражнения из тренировки
    global current_exercises, exercises, completed_count, completed_all_exercises

    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    workout_type = message.text.lower()
    picked_workout['type'] = workout_type  # Внесение данных в ключ "тип" тренировки
    gender = picked_workout['gender']  # Присваивание значения пола пользователя для удобства

    video_url = None  # Объявление переменной для ссылки на видео с тренировкой
    exercises = None  # Объявление переменной для множества упражнений в определенной тренировке

    completed_count = 0  # Кол-во выполненных упражнений

    # Отправка выбранного типа упражнений пользователю
    await message.answer(f'Отлично! Давайте приступим к типу упражнений: {workout_type}')

    # Определение типа видео, которое нужно отправить пользователю
    if picked_workout['experience'] == "без опыта" and gender == 'мужчина':
        if workout_type == 'на руки':
            video_url, exercises = male_arm_workout_video_beginner,  male_light_exercises['arms']
        elif workout_type == 'на ноги':
            video_url, exercises = male_leg_workout_video_beginner,  male_light_exercises['legs']
        elif workout_type == 'на пресс':
            video_url, exercises = male_abs_workout_video_beginner,  male_light_exercises['abs']
        else:
            video_url, exercises = male_whole_body_workout_video_beginner, male_light_exercises['whole_body']

    elif picked_workout['experience'] == "с опытом" and gender == 'мужчина':
        if workout_type == 'на руки':
            video_ur, exercises = male_arm_workout_video_advanced, male_advanced_exercises['arms']
        elif workout_type == 'на ноги':
            video_url, exercises = male_leg_workout_video_advanced, male_advanced_exercises['legs']
        elif workout_type == 'на пресс':
            video_url, exercises = male_abs_workout_video_advanced, male_advanced_exercises['abs']
        else:
            video_url, exercises = male_whole_body_workout_video_advanced, male_advanced_exercises['whole_body']

    elif picked_workout['experience'] == "без опыта" and gender == 'женщина':
        if workout_type == 'на руки':
            video_url, exercises = female_arm_workout_video_beginner, female_light_exercises['arms']
        elif workout_type == 'на ноги':
            video_url, exercises = female_leg_workout_video_beginner, female_light_exercises['legs']
        elif workout_type == 'на пресс':
            video_url, exercises = female_abs_workout_video_beginner, female_light_exercises['abs']
        else:
            video_url, exercises = female_whole_body_workout_video_beginner, female_light_exercises['whole_body']

    elif picked_workout['experience'] == "с опытом" and gender == 'женщина':
        if workout_type == 'на руки':
            video_ur, exercises = female_arm_workout_video_advanced, female_advanced_exercises['arms']
        elif workout_type == 'на ноги':
            video_url, exercises = female_leg_workout_video_advanced, female_advanced_exercises['legs']
        elif workout_type == 'на пресс':
            video_url, exercises = female_abs_workout_video_advanced, female_advanced_exercises['abs']
        else:
            video_url, exercises = female_whole_body_workout_video_advanced, female_advanced_exercises['whole_body']
    else:
        await message.answer("Похоже, нет доступных упражнений.")

    # При условии, что переменные для видео и упражнений имеют значения
    if video_url and exercises:
        # Отправка пользователю всех упражнений из тренировки (пронумерованных)
        await message.answer("Список упражнений:")
        for i, exercise_name in enumerate(exercises.keys(), start=1):
            await message.answer(f"{i}. {exercise_name}")

        # Отправка пользователю нужного видео
        await message.answer(f"Вот видео с тренировкой: {video_url}")
        current_exercises = list(exercises.keys())  # список текущих упражнений

        # Извлечение первого упражнения из current_exercises и присваивание его переменной exercise,
        # затем удаление упражнения из списка, чтобы упражнения сменяли друг друга
        exercise = current_exercises.pop(0)

        # Отправка пользователю названия упражнения с выбором выполнено/пропустить
        await message.answer(f"Упражнение: {exercise}", reply_markup=kb_complete_skip)


# Обработка нажатия на кнопки с выполнением/пропуском упражнения
@dp.message(or_f(F.text.lower() == "выполнено", F.text.lower() == "пропустить"))
async def handle_completion(message: types.Message):
    # Глобальные переменные:
    # 1. Список упражнений, предназначенных в текущей тренировке;
    # 2. Кол-во выполненных упражнений в текущей тренировке,
    # 3. Переменная, отслеживающая были ли выполнены все упражнения из тренировки
    global current_exercises, exercises, completed_count,  completed_all_exercises

    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    # Если все упражнения уже выполнены, игнор команды "Выполнено" и "Пропустить"
    if completed_all_exercises:
        await message.answer("Вы уже закончили тренировку")
        return  # Завершение функции, так как тренировка уже завершена

    # Получение "Выполнено" или "Пропустить" от пользователя
    action = message.text.lower()

    # Обработка выбранного действия
    if action == "выполнено":
        await message.answer("Отлично! Упражнение выполнено!")
        completed_count += 1  # Увеличение счетчика выполненных упражнений
    elif action == "пропустить":
        await message.answer("Упражнение пропущено")

    # Если есть еще упражнения в текущей тренировке, отправка следующего упражнения
    if current_exercises:
        # Извлечение первого упражнения из current_exercises и присваивание его переменной exercise,
        # затем удаление упражнения из списка, чтобы упражнения сменяли друг друга
        exercise = current_exercises.pop(0)

        # Отправка пользователю названия упражнения с выбором выполнено/пропустить
        await message.answer(f"Упражнение: {exercise}", reply_markup=kb_complete_skip)

    # Если упражнения закончились, вывод статистики выполнения тренировки
    else:
        completion_percentage = completed_count / len(exercises) * 100  # Вычисление процента выполненных упражнений

        # Отправка сообщения со статистикой выполнения тренировки
        await message.answer(f"Выполнено {completed_count}/{len(exercises)} упражнений ({completion_percentage:.2f}%)",
                             reply_markup=del_kbd)
        completed_all_exercises = True  # Установка флага, указывающего на завершение всех упражнений

        # Если все упражнения выполнены, поздравление пользователя
        if completion_percentage == 100:
            await message.answer("Вы справились со всеми упражнениями! Так держать!")


# Отправка пользователю мотивационных цитат
async def send_motivational_quote(message: types.Message):
    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    quote = random.choice(list(with_skip))  # Выбор случайной мотивационной цитаты

    # Отправка цитаты с выбором: отправка новой цитаты или переход к тренировке
    await message.answer(quote, reply_markup=kb_motivation)


@dp.message(Command("motivation"))  # обработка команды /motivation
async def handle_motivation_command(message: types.Message):
    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    #  Отправка сообщения с выбором: отправка новой цитаты или переход к тренировке
    await message.answer("Если вам нужно еще мотивации, нажмите 'Больше мотивации!', "
                         "или 'Приступим к тренировке!', если вы достаточно зарядились", reply_markup=kb_motivation)
    await send_motivational_quote(message)  # Отправка цитаты


# Обработка нажатия на кнопки для отправки другой цитаты/перехода к тренировке
@dp.message(or_f(F.text.lower() == "больше мотивации!", F.text.lower() == "приступим к тренировке!"))
async def handle_motivation_response(message: types.Message):
    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    if message.text.lower() == "больше мотивации!":
        await send_motivational_quote(message) # Отправка мотивационной цитаты
    else:
        await handle_workouts_command(message)  # Переход к выбору тренировки


# Обработка всех остальных сообщений
@dp.message()
async def handle_all_messages(message: types.Message):
    # Сохранение сообщений от пользователя в файл
    save_to_file(message.text)

    await message.answer('Вы всегда можете воспользоваться командами через "/"')


# Функция для сохранения реплик пользователя в файл
def save_to_file(message_text):
    # Присваивание файлу название
    filename = "user.txt"
    # Запись текста сообщения от пользователя в файл
    with open(filename, "a", encoding="utf-8") as file:
        # Разделение сообщений построчно
        file.write(message_text + "\n")


await main()  # Запуск основной функции асинхронного кода
```
